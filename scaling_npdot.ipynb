## 1. minmax sacler

```python
# minmax scaling : 최소값 0, 최대값 1로 정규화
# 비교적 직관적이지만 이상치에 영향을 많이 받음
```

```python
from sklearn.preprocessing import MinMaxScaler

# Normalization 최소값 0 / 최대값 1
scaler = MinMaxScaler()

# 행정동 컬럼 제외
minmax_trans = pd.DataFrame(scaler.fit_transform(df.iloc[:,1:]))
minmax_trans
```

## 2. robust scaler

```python
# Robust scaling :  중앙값 0, IQR 1이 되도록 스케일링하는 방법
# 최대, 최소값이 아닌 사분위수를 이용하기 때문에 이상치의 영향을 최소화 할 수 있다. 
```

```python
from sklearn.preprocessing import RobustScaler

scaler = RobustScaler()

# 행정동 컬럼 제외
robust_trans = pd.DataFrame(scaler.fit_transform(df.iloc[:,1:]))
robust_trans
```

## 3. 데이터프레임 재구성 (robust 스케일링 결과로 진행)

```python
# 컬럼명 다시 붙이기
a = df.columns
robust_trans.columns = a[1:]

# 행정동 인덱스로 넣기
robust_trans.index = df['행정동']

robust_trans
```

# 2. numpy를 이용한 매트릭스 계산 

```python
robust_trans.info()
```

```python
matrix_df = robust_trans.to_numpy() # dataframe을 numpy array로 변환 (타겟 행렬 구성)
matrix_df
```

```python
weight = np.random.random((51, 1)) # 가중치 행렬 구성 (난수발생)
weight
```

```python
output = np.dot(matrix_df,weight) # 타겟 행렬과 가중치 행렬의 dot product
output
```

```python
pd.DataFrame(output).sort_values(by=0, ascending=False) # out 데이터프레임 변환 후 정렬
```
